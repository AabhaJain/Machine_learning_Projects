# Under Sampling

Before proceeding with the Random Under Sampling technique we have to separate the orginal dataframe, for testing purposes, remember although we are splitting the data when implementing Random UnderSampling or OverSampling techniques, we want to test our models on the original testing set not on the testing set created by either of these techniques. The main goal is to fit the model either with the dataframes that were undersample and oversample (in order for our models to detect the patterns), and test it on the original testing set.

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import StratifiedKFold

x = data.drop('Class', axis = 1)
y = data['Class']

sk = StratifiedKFold(n_splits = 5, shuffle = False)

for train_index, test_index in sk.split(x, y):
    print("Train : ", train_index, "Test : ", test_index)
    original_x_train, original_x_test = x.iloc[train_index], x.iloc[test_index]
    original_y_train, original_y_test = y.iloc[train_index], y.iloc[test_index]
    
# We already have X_train and y_train for undersample data thats why I am using original to distinguish and to not overwrite these variables.

# Turn into an array

original_xtrain = original_x_train.values
original_xtest = original_x_test.values
original_ytrain = original_y_train.values
original_ytest = original_y_test.values

# See if both the train and test label distribution are similarly distributed

train_unique_label, train_counts_label = np.unique(original_ytrain, return_counts = True)
test_unique_label, test_counts_label = np.unique(original_ytest, return_counts = True)
print('-'*130)

print("Train Unique Label: ", train_unique_label)
print("Train count Label: ", train_counts_label)
print("Test Unique Label: ", test_unique_label)
print("Test count Label: ", test_counts_label, end = "")

print("\n")
print('Label Distributions: \n')
print(train_counts_label/ len(original_ytrain))
print(test_counts_label/ len(original_ytest))
```

# Anomaly Detection

In this section is to remove "extreme outliers" from features that have a high correlation with our classes. This will have a positive impact on the accuracy of our models.

```python
from scipy.stats import norm

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20,6))

v14_fraud = new[new['Class'] == 1]['V14'].values
sns.distplot(v14_fraud, color = 'purple', fit = norm, ax = ax1)
ax1.set_title("V14 Fraud distribution")

v12_fraud = new[new['Class'] == 1]['V12'].values
sns.distplot(v12_fraud, color = 'green', fit = norm, ax = ax2)
ax2.set_title("V12 Fraud distribution")

v10_fraud = new[new['Class'] == 1]['V10'].values
sns.distplot(v10_fraud, color = '#FB8861', fit = norm, ax = ax3)
ax3.set_title("V10 Fraud distribution")
```

V14 is the only feature that has most guassian feature comapred to V12, V10

```python
# Removing outliers from V14 (highest negative correlated feature with the Class label)

fraud = data[data['Class'] == 1]['V14'].values
q1, q3 = np.percentile(fraud, 25) , np.percentile(fraud, 75)
print("Quartile 3 :", q3, "| Quartile 1 :",q1)

v14_iqr = q3 - q1
print("iqr : {}" .format(v14_iqr))

# As the data points which fall below Q1 – 1.5 IQR or above Q3 + 1.5 IQR are outliers.

v14_cutoff = 1.5 * v14_iqr 
lower = q1 - v14_cutoff
upper = q3 + v14_cutoff

print("Cut Off :", v14_cutoff) 
print("V14 lower :", lower)
print("V14 Upper :", upper)

outliers = [x for x in fraud if (x < lower) or (x > upper)]
print("No. of outliers in V14 :", len(outliers))
print("V14 outliers :", outliers)

# Drop these outliers from the data

new = new.drop(new[(new['V14'] > upper) | (new['V14'] < lower)].index)
print(new.shape)
```

# Dimensionality Reduction and Clustering

T-sne and PCA both are used for dimensionality reduction. PCA preserves the variance in the data, whereas t-SNE preserves the relationships between data points in a lower-dimensional space. PCA is a linear Dimensionality reduction technique while t-sne is a non-linear Dimensionality reduction technique.


```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

x = new.drop('Class', axis = 1)
y = new['Class']

x_tsne = TSNE(n_components = 2, random_state = 40).fit_transform(x.values)
print("Original x shape :", x.shape)
print("x reduced tsne shape :", x_tsne.shape)

x_pca = PCA(n_components = 2, random_state = 40).fit_transform(x.values)
#print("Original x shape :", len(x))
print("x reduced pca shape :", x_pca.shape)
```

```python
# Use grid Search cv to find the best parameters

from sklearn.model_selection import GridSearchCV

# logistic Regression
log_reg_paras = {'penalty' : ['l1', 'l2'], 'C' : [0.01, 0.001, 0.1, 1, 10, 100]}
grid_log_reg = GridSearchCV(LogisticRegression(), log_reg_paras, scoring = "accuracy" )
grid_log_reg.fit(x_train, y_train)

log_reg = grid_log_reg.best_estimator_
print("Logistic Regression :", grid_log_reg.best_estimator_)

# KNN
knn_paras = {'n_neighbors' : range(0, 21), 'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}
grid_knn = GridSearchCV(KNeighborsClassifier(), knn_paras )
grid_knn.fit(x_train, y_train)

knears_neighbours = grid_knn.best_estimator_
print("KNN :", grid_knn.best_estimator_)

# Support Vector Classifier
svc_params = {'C': [0.5, 0.7, 0.9, 1], 'kernel': ['rbf', 'poly', 'sigmoid', 'linear']}
grid_svc = GridSearchCV(SVC(), svc_params)
grid_svc.fit(x_train, y_train)

svc = grid_svc.best_estimator_
print("SVC :", grid_svc.best_estimator_)

# Decision Tree Classifier
tree_params = {"criterion": ["gini", "entropy"], "max_depth": list(range(2,4,1)), "min_samples_leaf": list(range(5,7,1))}
grid_tree = GridSearchCV(DecisionTreeClassifier(), tree_params)
grid_tree.fit(x_train, y_train)

tree_clf1 = grid_tree.best_estimator_
print("Descision Tree Classifier :", grid_tree.best_estimator_)

# Random Forest Classifier
param = {'max_depth' : list(range(1,10,1)), 'criterion' : ['gini', 'entropy'], 'n_estimators' : [110,120,130,140,150]}
grid_search = GridSearchCV(RandomForestClassifier() , param_grid = param)
grid_search.fit(x_train, y_train)

tree_clf2 = grid_tree.best_estimator_
print("Random Forest Classifier :", grid_search.best_estimator_)
```

```python
# Let's plot logistic Regression learning curve

from sklearn.model_selection import learning_curve

def plot_learning_curve(estimator1, estimator2, estimator3, estimator4, x, y, ylim = None, cv = None, scoring = 'accuracy',
                        n_jobs = 1, train_sizes = np.linspace(0.01, 1.0, 5), sharey = True):
    
    
    f, ((ax1, ax2) , (ax3, ax4)) = plt.subplots(2,2, figsize = (29,15))
    
    if ylim is not None:
        plt.ylim(*ylim)
        
    train_size = [100, 200, 300, 400,500, 600]    
    # estimator 1
    train_sizes, train_scores, test_scores = learning_curve(estimator1, x, y, cv = cv, scoring = scoring, 
                                                            n_jobs = n_jobs, train_sizes = train_sizes)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)

    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    ax1.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha = 0.1, color = "#ff9124")
    ax1.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha = 0.1, color = "#2492ff")
    
    ax1.plot(train_sizes, train_mean, 'o-', color="#ff9124",  label="Training score")
    ax1.plot(train_sizes, test_mean, 'o-', color="#2492ff", label="Cross-validation score")

    ax1.set_title("Logistic Regresion Learning Curve", fontsize = 16)
    ax1.set_xlabel("Training Set Size")
    ax1.set_ylabel("Accuracy Score")
    ax1.legend(loc="best")
    ax1.grid(True)
    
    # estimator 2
    train_sizes, train_scores, test_scores = learning_curve(estimator2, x, y, cv = cv, scoring = scoring, 
                                                            n_jobs = n_jobs, train_sizes = train_sizes)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)

    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    ax2.plot(train_sizes, train_mean, 'o-', color="#ff9124",  label="Training score")
    ax2.plot(train_sizes, test_mean, 'o-', color="#2492ff", label="Cross-validation score")

    ax2.fill_between(train_sizes, train_mean - train_std, train_mean + train_std,alpha = 0.1, color="#ff9124")
    ax2.fill_between(train_sizes, test_mean - test_std, test_mean + test_std,alpha = 0.1, color="#2492ff")

    ax2.set_title("KNN Learning Curve", fontsize = 16)
    ax2.set_xlabel("Training Set Size")
    ax2.set_ylabel("Accuracy Score")
    ax2.legend(loc="best")
    ax2.grid(True)
    
    # estimator3
    train_sizes, train_scores, test_scores = learning_curve(estimator3, x, y, cv = cv, scoring = scoring, 
                                                            n_jobs = n_jobs, train_sizes = train_sizes)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)

    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    ax3.plot(train_sizes, train_mean, 'o-', color="#ff9124",  label="Training score")
    ax3.plot(train_sizes, test_mean, 'o-', color="#2492ff", label="Cross-validation score")

    ax3.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha = 0.1,color="#ff9124")
    ax3.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha = 0.1,color="#2492ff")

    ax3.set_title("SVC Learning Curve", fontsize = 16)
    ax3.set_xlabel("Training Set Size")
    ax3.set_ylabel("Accuracy Score")
    ax3.legend(loc="best")
    ax3.grid(True)
    
    # estimator4 
    train_sizes, train_scores, test_scores = learning_curve(estimator4, x, y, cv = cv, scoring = scoring, 
                                                            n_jobs = n_jobs, train_sizes = train_sizes)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)

    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    ax4.plot(train_sizes, train_mean, 'o-', color="#ff9124",  label="Training score")
    ax4.plot(train_sizes, test_mean, 'o-', color="#2492ff", label="Cross-validation score")

    ax4.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha = 0.1, color="#ff9124")
    ax4.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha = 0.1, color="#2492ff")

    ax4.set_title("Descision Tree Learning Curve", fontsize = 16)
    ax4.set_xlabel("Training Set Size")
    ax4.set_ylabel("Accuracy Score")
    ax4.legend(loc="best")
    ax4.grid(True)
    
```
